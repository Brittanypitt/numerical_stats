```python
#3.1

P=[0.2,0.3,0.1,0.4]     #will save the probabilities of each height

E_height=np.sum(195*P[0]+180*P[1]+168*P[2]+170*P[3])
print('The expected value is:',E_height)
Variance=-E_height**2+np.sum(195**2*P[0]+180**2*P[1]+168**2*P[2]+170**2*P[3])
print('The variance is:',Variance)
```

$Z=\int_{-2}^{2}(-x^2+4)dx=\int_{-2}^{2}(-x^2+4)dx=(\frac{-x^3}{3}+4x)|_{-2}^{2}=[-\frac{8}{3}+4\cdot2]-[\frac{-8}{3}+4\cdot(-2)]$=16
$E(x)=\frac{\int_{-2}^{2}x*(-x^2+4)dx}{Z}=\frac{\int_{-2}^{2}(-x^3+4x)dx}{16}=\frac{(\frac{-x^4}{4}+2x^2)|_{-2}^{2}}{16}=\frac{[-\frac{16}{4}+2\cdot2^2]-[-\frac{16}{4}+2\cdot(-2)^2]}{16}=0$
